```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('multiple_linear_regression_dataset.csv', encoding = 'latin1')
```

```python
x = data [['age' , 'experience']].values 
y = data[['income']].values
```

```python
from sklearn.model_selection import train_test_split as tts 
x_train, x_test, y_train, y_test = tts(
    x, y, test_size=0.3, random_state=42
)
```

```python
x_mean = x_train.mean(axis=0)
x_std = x_train.std(axis=0)

x_train = (x_train - x_mean) / x_std
x_test = (x_test - x_mean) / x_std
```

```python
x_train = np.c_[np.ones(x_train.shape[0]), x_train]
x_test = np.c_[np.ones(x_test.shape[0]), x_test]
```

```python
learning_rate = 0.01
epochs = 1000
weights = np.zeros(x_train.shape[1])
```

```python
for epoch in range(epochs):
    for i in range(len(x_train)):
        y_pred = np.dot(x_train[i], weights)
        error = y_train[i] - y_pred
        weights = weights + learning_rate * error * x_train[i]
```

```python
y_test_pred = np.dot(x_test, weights)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_test_pred)
r2 = r2_score(y_test, y_test_pred)

print("MSE:", mse)
print("RÂ² Score:", r2*100, "%")
```
